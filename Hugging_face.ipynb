```python
from langchain_huggingface import HuggingFacePipeline
from transformers import AutoModelForCausalLM, AutoTokenizer, pipeline

model_id = "gpt2"
model = AutoModelForCausalLM.from_pretrained(model_id)
tokenizer = AutoTokenizer.from_pretrained(model_id)

# Ensure that pipeline is a Hugging Face pipeline object
pipeline = pipeline(
    "text-generation",
    model=model, # Add the model back
    tokenizer=tokenizer,
    max_new_tokens=100)
huggingface_llm = HuggingFacePipeline(pipeline=pipeline)

huggingface_llm.invoke("langchain")
```
